Acura model and Sales

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

all_car_sales=pd.read_csv("Car_sales.csv")
manufacturer_sales = all_car_sales.groupby("Manufacturer")["Sales_in_thousands"].sum().reset_index()

plt.figure(figsize=(12,7))
plt.bar(manufacturer_sales["Manufacturer"],manufacturer_sales["Sales_in_thousands"],color="orange")
plt.xlabel("Manufacturer")
plt.ylabel("Sales in Thousands")
plt.title("Total Sales by Manufacturer")
plt.xticks(rotation=45,ha="right")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,8))
colors=plt.get_cmap("YlOrRd")(np.linspace(0.3,0.6,len(manufacturer_sales)))
plt.pie(manufacturer_sales["Sales_in_thousands"],labels=manufacturer_sales["Manufacturer"],colors=colors,autopct="%1.1f%%",startangle=90)
plt.title("Sales Distribution by Manufacturer")
plt.axis("equal")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

car_sales = pd.read_csv('Car_sales.csv')
aura_sales = car_sales[car_sales['Manufacturer'] == 'Acura']

plt.figure(figsize=(6, 4))
plt.bar(aura_sales['Model'], aura_sales['Sales_in_thousands'])
plt.xlabel('Car Model (Acura)')
plt.ylabel('Sales in Thousands')
plt.title('Sales in Thousands for Acura Models (Bar Chart)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(aura_sales)))
plt.pie(aura_sales['Sales_in_thousands'], labels=aura_sales['Model'], colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Acura Model (Pie Chart)')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
